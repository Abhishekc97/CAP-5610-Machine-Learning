```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
from IPython.display import Image
from sklearn import tree
```

```python
train_data = {'Home': [1, 1, 0, 0, 1, 0],
             'AP25':  [0, 1, 0, 0, 0, 0],
             'NBC':   [1, 1, 0, 0, 1, 0],
             'ESPN':  [0, 0, 1, 0, 0, 0],
             'FOX':   [0, 0, 0, 1, 0, 0],
             'ABC':   [0, 0, 0, 0, 0, 1],
             'Win':   [1, 0, 1, 1, 1, 1]
        }
test_data = {'Home': [1, 1, 1, 0, 1, 0],
             'AP25': [1, 0, 0, 1, 0, 1],
             'NBC':   [1, 1, 1, 0, 1, 0],
             'ESPN':  [0, 0, 0, 0, 0, 0],
             'FOX':   [0, 0, 0, 0, 0, 0],
             'ABC':   [0, 0, 0, 1, 0, 1],
        }
```

```python
df_train = pd.DataFrame(data=train_data)
df_test = pd.DataFrame(data=test_data)
```

```python
feature_cols = ['Home', 'AP25', 'NBC', 'ESPN', 'FOX', 'ABC']

x_train = df_train[feature_cols]
y_train = df_train.Win

x_test = df_test[feature_cols]
# y_test = df_test.Win

clf = DecisionTreeClassifier(criterion="entropy")
clf = clf.fit(x_train, y_train)

y_pred = clf.predict(x_test)

# print('x_train',x_train)
# print('x_test',x_test)

print('y_pred:', y_pred)
tree.plot_tree(clf.fit(x_train, y_train))
```

```python
clf2 = DecisionTreeClassifier()
clf2 = clf2.fit(x_train, y_train)

y_pred = clf2.predict(x_test)

tree.plot_tree(clf2.fit(x_train, y_train))
```

```python
df_train.to_csv('q4aTrain.csv', sep='\t', encoding='utf-8', index=False)
df_test.to_csv('q4aTest.csv', sep='\t', encoding='utf-8', index=False)
```
